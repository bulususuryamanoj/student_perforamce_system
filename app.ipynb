<h1><b>Stage-3 : DataPreprocessing</b></h1>
In this Step We do <ol><li>Data Cleaning</li><li>Data Transforamation<li>Data Integration</li></ol>
<p>For our project we are using only using Data Transformation and cleaning because, <ul><li>The data is enough for our basic analysis, no need <b> Data integration</b></li></ul>

-> Data Cleaning : Removing Some Unwanted Columns like <ol><li>Student_ID</li><li>Paren_Education_Level</li></ul>

```python
data = data.drop(['Student_ID','Parental_Education_Level'],axis=1)
```

```python
print(data.head(5))
```

using the Simple replace() method to convert the data from categorical to numerical <br>
The Categorical Column are : 
<ol>
<li>Gender</li>
<li>Internet_Access_at_Home</li>
<li>Extracurriclar_Activites</li>
<li>Pass_Fail</li>
</ol>

```python
data['Gender'].replace({"Male":1,"Female":0},inplace=True)
#inplace = True -> Replaces the transformed data with original data in the dataset
data['Internet_Access_at_Home'].replace({"Yes":1,"No":0},inplace=True)
data['Extracurricular_Activities'].replace({"Yes":1,"No":0},inplace=True)
data['Pass_Fail'].replace({"Pass":1,"Fail":0},inplace=True)
```

```python
data.info()
print(data.head(5))
```

We can see all the data is in the numerical form (int/float)


<b><h1>Stage-4 : Model Building using Scikit-Learn </h1><b>


<h3>Identifying Depenedent(y) and Independent(x) Variables </h3>


```python
x = data.iloc[:,:7] # [all rows, 0 to 6 columns]
print("Independent Variable:\n",x)
y = data['Pass_Fail']
print()
print("Dependent Variable:\n",y)
```

Splitting data into training and testing data using train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

Using Decision Tree to build the model


```python
from sklearn.tree import DecisionTreeClassifier
model  = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

model Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import accuracy_score
acs = accuracy_score(y_test,y_pred)
print(acs)
```
